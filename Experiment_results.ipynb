```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,150)
amp = pd.read_csv('/content/resnet_amp.csv')
half = pd.read_csv('/content/resnet_half.csv')
single = pd.read_csv('/content/resnet_single.csv')

plt.plot(x, amp[" test_avg_loss"], label="amp")
plt.plot(x, half[" test_avg_loss"], label="half")
plt.plot(x, single[" test_avg_loss"], label="single")
plt.legend()
plt.ylabel("Test Loss")
plt.xlabel("Epochs")
plt.show()
```

```python
plt.plot(x, amp[" test_acc"], label="amp")
plt.plot(x, half[" test_acc"], label="half")
plt.plot(x, single[" test_acc"], label="single")
plt.legend()
plt.ylabel("Test Accuracy")
plt.xlabel("Epochs")
plt.show()
```

```python
amptime=amp[' train_time'].mean()
halftime=half[' train_time'].mean()
singletime=single[' train_time'].mean()

modes = ['half', 'amp', 'single']
time = [halftime,amptime,singletime]
plt.bar(modes, time, width = 0.4)
plt.ylabel("Average training time per epoch (s)")
plt.xlabel("modes")
plt.show()
```
