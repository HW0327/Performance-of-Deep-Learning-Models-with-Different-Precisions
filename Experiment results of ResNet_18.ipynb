```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,150)
amp = pd.read_csv('/content/resnet_amp.csv')
half = pd.read_csv('/content/resnet_half.csv')
single = pd.read_csv('/content/resnet_single.csv')
ssdamp = pd.read_csv('/content/SSD_amp.csv')
ssdsingle = pd.read_csv('/content/SSD_single.csv')
ssdhalf = pd.read_csv('/content/SSD_half.csv')

plt.plot(x, amp[" test_avg_loss"], label="amp")
plt.plot(x, half[" test_avg_loss"], label="half")
plt.plot(x, single[" test_avg_loss"], label="single")
plt.legend()
plt.ylabel("Test Loss")
plt.xlabel("Epochs")
plt.show()
```

```python
plt.plot(x, amp[" test_acc"], label="amp")
plt.plot(x, half[" test_acc"], label="half")
plt.plot(x, single[" test_acc"], label="single")
plt.legend()
plt.ylabel("Test Accuracy")
plt.xlabel("Epochs")
plt.show()
```

```python
amptime=amp[' train_time'].mean()
halftime=half[' train_time'].mean()
singletime=single[' train_time'].mean()

modes = ['half', 'amp', 'single']
time = [halftime,amptime,singletime]
plt.bar(modes, time, width = 0.4)
plt.ylabel("Average training time per epoch (s)")
plt.xlabel("modes")
plt.show()
print("half mode:",halftime,"s")
print("amp mode:",amptime, "s")
print("single mode:",singletime, "s")
```

```python
z=np.arange(0,58)
plt.plot(z, ssdamp[" Loss"], label="amp")
plt.plot(z, ssdhalf[" Loss"], label="half")
plt.plot(z, ssdsingle[" Loss"], label="single")
plt.legend()
plt.ylabel("Test Loss")
plt.xlabel("Epochs")
plt.show()
```

```python
plt.plot(z, ssdamp[" mAP"], label="amp")
plt.plot(z, ssdhalf[" mAP"], label="half")
plt.plot(z, ssdsingle[" mAP"], label="single")
plt.legend()
plt.ylabel("mAP")
plt.xlabel("Epochs")
plt.show()
```

```python
ssdamptime=ssdamp[' Batch Time'].mean()
ssdhalftime=ssdhalf[' Batch Time'].mean()
ssdsingletime=ssdsingle[' Batch Time'].mean()

modes = ['half', 'amp', 'single']
time = [ssdhalftime,ssdamptime,ssdsingletime]
plt.bar(modes, time, width = 0.4)
plt.ylabel("Average training time per epoch (hr)")
plt.xlabel("modes")
plt.show()
print("half mode:",ssdhalftime,"hour")
print("amp mode:",ssdamptime, "hour")
print("single mode:",ssdsingletime, "hour")
```
